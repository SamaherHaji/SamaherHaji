```python
# drop unrelevent colunms: URLs and description colunms, are unlikely to contribute to analysis
# drop county colunm bcs it have no values (completely null)
to_drop = ['url', 'region_url', 'image_url', 'description', 'county', 'size']
df.drop(columns = to_drop, inplace=True)
```

```python
#Impute missing values for important columns
#impute missing values for numerical values with madian
numeric_cols = ['year', 'odometer']
for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Impute missing categorical values with mode
categorical_cols = ['manufacturer', 'model', 'condition', 'cylinders', 'fuel', 'title_status', 'drive', 'type', 'paint_color', 'transmission']
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Impute posting_date with forward fill
df['posting_date'] = df['posting_date'].fillna(method='ffill').fillna(method='bfill')
```

```python
#correct data types: posting_data to datetime and year to integer
df['posting_date'] = pd.to_datetime(df['posting_date'], utc=True)
df['year'] = df['year'].astype(int)
```
